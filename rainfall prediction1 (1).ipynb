```python
import pandas as pd
```

```python
full_data = pd.read_csv("D:\\dataset\\weatherAUS.csv")
```

```python
# it is used to display the first few rows of a DataFrame to quickly inspect its contents in Python, often for data exploration
#and understanding.
full_data.head()
```

```python
# This code converts a 'Date' column to datetime format, extracts year, month, and day components into separate columns, 
# then drops the original 'Date' column, resulting in a DataFrame with added time-related features for analysis.

full_data['Date'] = pd.to_datetime(full_data['Date'])
full_data['year'] = full_data['Date'].dt.year
full_data['month'] = full_data['Date'].dt.month
full_data['day'] = full_data['Date'].dt.day
full_data.drop(['Date'], axis = 1,inplace=True) 
full_data.head()
```

```python
#is used to retrieve the dimensions (number of rows and columns) of a DataFrame or array in Python.
full_data.shape
```

```python
# is used to display concise summary information about a DataFrame in Python, including the data types,
# non-null values, and memory usage, aiding in initial data assessment and quality checks.
full_data.info()
```

```python
#This code replaces categorical values in 'RainToday' and 'RainTomorrow' columns with numerical 
#equivalents (0 for 'No', 1 for 'Yes') for compatibility with numerical analysis and modeling.

full_data['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
full_data['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
full_data.head()
```

```python
#This code generates a bar plot visualizing the distribution of 'RainTomorrow' indicator values (No: 0, Yes: 1) 
#in an imbalanced dataset, emphasizing proportions with rounded values annotated above each bar.

import matplotlib.pyplot as plt
fig = plt.figure(figsize = (20,5))
ax=full_data.RainTomorrow.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('RainTomorrow Indicator No(0) and Yes(1) in the Imbalanced Dataset')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(),2)), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.show()
```

```python
# This code oversamples the minority class ('RainTomorrow' == 1) to balance it with the majority class ('RainTomorrow' == 0) 
#in the dataset, then creates a bar plot to visualize the distribution of the balanced classes, aiding in the assessment of 
#the class imbalance resolution.

from sklearn.utils import resample

no = full_data[full_data.RainTomorrow == 0]
yes = full_data[full_data.RainTomorrow == 1]
yes_oversampled = resample(yes, replace=True, n_samples=len(no), random_state=42)
oversampled = pd.concat([no, yes_oversampled])

fig = plt.figure(figsize = (20,5))
ax=oversampled.RainTomorrow.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('RainTomorrow Indicator No(0) and Yes(1) after Oversampling (Balanced Dataset)')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(),2)), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.show()
```

```python
# Missing Data Pattern in Training Data
#This code uses Seaborn to create a heatmap visualization of missing values (null or NaN) in the 'oversampled' DataFrame,
#where each cell represents the presence (blue) or absence (white) of missing data for different features, aiding in 
#identifying data gaps and patterns.

import seaborn as sns
plt.figure(figsize = (20,5))
sns.heatmap(oversampled.isnull(), cbar=False, cmap='PuBu')
plt.show()
```

```python
# This code calculates and displays the count and percentage of missing values for each column in the DataFrame "oversampled," 
# helping to identify which columns have the most missing data and their corresponding proportions, which is essential for data 
# quality assessment and preprocessing decisions.

total = oversampled.isnull().sum().sort_values(ascending=False)
percent = (oversampled.isnull().sum()/oversampled.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing.head()
```

```python
#  is used to identify and select columns in a DataFrame (oversampled) that have data type 'object', which typically
# corresponds to categorical variables, for further analysis or processing.
oversampled.select_dtypes(include=['object']).columns
```

```python
# Impute categorical var with Mode
# his code fills missing values in specific columns ('Location', 'WindGustDir', 'WindDir9am', and 'WindDir3pm') of the 
# DataFrame 'oversampled' with the most frequent value (mode) present in each respective column, ensuring data completeness
# for further analysis or modeling.

oversampled['Location'] = oversampled['Location'].fillna(oversampled['Location'].mode()[0])
oversampled['WindGustDir'] = oversampled['WindGustDir'].fillna(oversampled['WindGustDir'].mode()[0])
oversampled['WindDir9am'] = oversampled['WindDir9am'].fillna(oversampled['WindDir9am'].mode()[0])
oversampled['WindDir3pm'] = oversampled['WindDir3pm'].fillna(oversampled['WindDir3pm'].mode()[0])
```

```python
# Convert categorical features to continuous features with Label Encoding
# This code snippet utilizes LabelEncoder from the sklearn.preprocessing module to convert categorical (string) data 
# in selected columns of the 'oversampled' DataFrame into numerical values. This transformation is often necessary for machine
# learning algorithms that require numerical inputs, enabling the model to process and learn from the categorical data 
# effectively. The dictionary 'lencoders' stores the fitted LabelEncoder instances for potential later use.


from sklearn.preprocessing import LabelEncoder
lencoders = {}
for col in oversampled.select_dtypes(include=['object']).columns:
    lencoders[col] = LabelEncoder()
    oversampled[col] = lencoders[col].fit_transform(oversampled[col])
```

```python
# This code snippet uses the Multiple Imputation by Chained Equations (MICE) technique to impute missing values in the 
# "oversampled" DataFrame by iteratively estimating missing values based on observed values in other columns, effectively 
# filling in gaps in the data for improved predictive modeling or analysis. The warnings are temporarily suppressed, and the
# imputed data is stored in a new DataFrame called "MiceImputed."

import warnings
warnings.filterwarnings("ignore")
# Multiple Imputation by Chained Equations
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
MiceImputed = oversampled.copy(deep=True) 
mice_imputer = IterativeImputer()
MiceImputed.iloc[:, :] = mice_imputer.fit_transform(oversampled)
```

```python
# Detecting outliers with IQR
# This code calculates the interquartile range (IQR) for each column in the DataFrame MiceImputed, which is a measure of the 
# spread or dispersion of the data. It is used for detecting potential outliers or extreme values within the dataset by 
# comparing values that fall outside the range defined by Q1 - 1.5 * IQR and Q3 + 1.5 * IQR.

Q1 = MiceImputed.quantile(0.25)
Q3 = MiceImputed.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Removing outliers from the dataset
# This code filters out rows in the DataFrame MiceImputed that have values beyond 1.5 times the interquartile range (IQR) from 
#the median, removing potential outliers, and then shows the new shape of the DataFrame.

MiceImputed = MiceImputed[~((MiceImputed < (Q1 - 1.5 * IQR)) |(MiceImputed > (Q3 + 1.5 * IQR))).any(axis=1)]
MiceImputed.shape
```

```python
# Standardizing data
# This code snippet standardizes the data using Min-Max scaling, transforming the values in a DataFrame (MiceImputed) to a 
# normalized range between 0 and 1 while preserving their relationships.

from sklearn import preprocessing
r_scaler = preprocessing.MinMaxScaler()
r_scaler.fit(MiceImputed)
modified_data = pd.DataFrame(r_scaler.transform(MiceImputed), index=MiceImputed.index, columns=MiceImputed.columns)
```

```python
# Feature Importance using Filter Method (Chi-Square)
# This code snippet is performing feature selection using the chi-squared (Ï‡Â²) statistical test, which helps identify 
# the most influential features for predicting the 'RainTomorrow' variable. It selects the top 10 features from the 
#'modified_data' DataFrame (excluding the 'RainTomorrow' column) that have the highest association with the target variable. 
# The selected feature column names are then printed to understand which features have the strongest relationship with 
# predicting whether it will rain tomorrow.

from sklearn.feature_selection import SelectKBest, chi2
X = modified_data.loc[:,modified_data.columns!='RainTomorrow']
y = modified_data[['RainTomorrow']]
selector = SelectKBest(chi2, k=10)
selector.fit(X, y)
X_new = selector.transform(X)
print(X.columns[selector.get_support(indices=True)])
```

```python
# This code performs feature selection using a Random Forest Classifier to identify important features (predictors) for
# predicting the 'RainTomorrow' target variable. The SelectFromModel class is used to select features based on their 
# importance scores, and the code then prints the selected features and the corresponding feature importances from the 
# Random Forest model.

from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as rf

X = MiceImputed.drop('RainTomorrow', axis=1)
y = MiceImputed['RainTomorrow']
selector = SelectFromModel(rf(n_estimators=100, random_state=0))
selector.fit(X, y)
support = selector.get_support()
features = X.loc[:,support].columns.tolist()
print(features)
print(rf(n_estimators=100, random_state=0).fit(X,y).feature_importances_)
```

```python
# This code snippet is preparing a dataset for machine learning by selecting specific features and the target variable 
# ('RainTomorrow'), splitting the data into training and testing sets, and then normalizing the features using standard 
# scaling to ensure consistent and appropriate ranges for the machine learning model.

features = MiceImputed[['Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustDir', 
                       'WindGustSpeed', 'WindDir9am', 'WindDir3pm', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 
                       'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 
                       'RainToday']]
target = MiceImputed['RainTomorrow']

# Split into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42,
                                                    shuffle=True, stratify=target)

# Normalize Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# This function is used to plot the Receiver Operating Characteristic (ROC) curve, which visualizes the performance of a binary 
# classification model by showing the trade-off between the true positive rate (sensitivity) and false positive rate as the 
# classification threshold changes. The curve is plotted with the model's performance in orange against a diagonal reference 
# line (dashed blue) representing random guessing.

def plot_roc_cur(fper, tper):  
    plt.plot(fper, tper, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
```

```python
# This Python code defines a function run_model that trains a given machine learning model on training data 
# (X_train and y_train), evaluates its performance on test data (X_test and y_test), and provides various performance metrics 
# (accuracy, ROC AUC, Cohen's Kappa), displays a classification report, and plots an ROC curve to assess the model's quality. 
# This function streamlines the process of training, evaluating, and visualizing machine learning models while presenting 
# important metrics and insights for analysis and comparison.


import time
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, roc_auc_score, cohen_kappa_score, roc_curve, classification_report

def run_model(model, X_train, y_train, X_test, y_test, verbose=True):
    t0 = time.time()
    if verbose == False:
        model.fit(X_train, y_train, verbose=0)
    else:
        model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    coh_kap = cohen_kappa_score(y_test, y_pred)
    time_taken = time.time() - t0
    print("Accuracy = {}".format(accuracy))
    print("ROC Area under Curve = {}".format(roc_auc))
    print("Cohen's Kappa = {}".format(coh_kap))
    print("Time taken = {}".format(time_taken))
    print(classification_report(y_test, y_pred, digits=5))

    probs = model.predict_proba(X_test)
    probs = probs[:, 1]
    fper, tper, thresholds = roc_curve(y_test, probs)
    plt.figure()
    plt.plot(fper, tper, color='darkorange', lw=2, label='ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.show()

    
    
    plt.show()

    return model, accuracy, roc_auc, coh_kap, time_taken

# Rest of your code remains unchanged

# Call the run_model function with your model and data
```

```python
# Logistic Regression
# Logistic Regression Model: Logistic Regression is a statistical method used for binary classification problems. \
# It models the relationship between a binary outcome variable and one or more predictor variables. The logistic regression 
# algorithm tries to find the best-fitting line (or hyperplane) that separates the classes in the input data.

# Overall, this code snippet represents the process of configuring, training, and evaluating a logistic regression model
#  using L1 regularization and the liblinear solver. The choice of hyperparameters and solver can impact the model's 
# performance, and the provided run_model function likely helps streamline the process of training and evaluating various
# machine learning models.

from sklearn.linear_model import LogisticRegression

params_lr = {'penalty': 'l1', 'solver':'liblinear'}

model_lr = LogisticRegression(**params_lr)
model_lr, accuracy_lr, roc_auc_lr, coh_kap_lr, tt_lr = run_model(model_lr, X_train, y_train, X_test, y_test)
```

```python
# Decision Tree
# Explanation: Decision Tree classifiers are a type of machine learning model that are particularly useful for classification
# tasks. They create a tree-like structure where each internal node represents a decision based on a specific feature, 
# and each leaf node represents a class label. The hyperparameters you've set, such as max_depth and max_features, control the 
# complexity of the tree and can impact factors like overfitting and model performance.

# By using Decision Trees, you're leveraging a model that can capture complex decision boundaries and relationships within the 
# data. The hyperparameter settings, like max_depth and max_features, help tailor the model's behavior to strike a balance
# between capturing fine details and avoiding excessive complexity. The evaluation metrics calculated through the run_model 
# function give you insights into how well the model performs on your specific dataset.


from sklearn.tree import DecisionTreeClassifier

params_dt = {'max_depth': 16,
             'max_features': "sqrt"}

model_dt = DecisionTreeClassifier(**params_dt)
model_dt, accuracy_dt, roc_auc_dt, coh_kap_dt, tt_dt = run_model(model_dt, X_train, y_train, X_test, y_test)
```

```python
# Neural Network
# In summary, this code snippet sets up a neural network classifier with specific parameters, trains it on training data, 
# evaluates its performance on testing data, and collects various evaluation metrics. The function run_model seems to handle 
# the training and evaluation process, but its specific implementation is not provided in the given code snippet.

from sklearn.neural_network import MLPClassifier

params_nn = {'hidden_layer_sizes': (30,30,30),
             'activation': 'logistic',
             'solver': 'lbfgs',
             'max_iter': 500}

model_nn = MLPClassifier(**params_nn)
model_nn, accuracy_nn, roc_auc_nn, coh_kap_nn, tt_nn = run_model(model_nn, X_train, y_train, X_test, y_test)
```

```python
# Random Forest

# In short, the code is setting up and using a Random Forest classifier with specific hyperparameters for a classification task.
# The run_model function is employed to handle the training, testing, and evaluation of the model. The goal is to achieve
# accurate predictions on the given dataset using the Random Forest algorithm.

from sklearn.ensemble import RandomForestClassifier

params_rf = {'max_depth': 16,
             'min_samples_leaf': 1,
             'min_samples_split': 2,
             'n_estimators': 100,
             'random_state': 42}

model_rf = RandomForestClassifier(**params_rf)
model_rf, accuracy_rf, roc_auc_rf, coh_kap_rf, tt_rf = run_model(model_rf, X_train, y_train, X_test, y_test)
```

```python
# This code snippet is an example of using machine learning classifiers to plot decision boundaries for binary classification.
# It uses various classifiers, including Logistic Regression, Decision Tree, Neural Network, and Random Forest, along with an 
# ensemble classifier. The goal is to visualize how these classifiers separate different classes in a three-dimensional
# feature space.


# In short, this code aims to visually compare and contrast how different classifiers separate two classes in a
# three-dimensional feature space. It provides insight into the decision boundaries created by each classifier, which can help
# understand their behavior and performance on the given dataset.


import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import itertools
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from mlxtend.classifier import EnsembleVoteClassifier
from mlxtend.plotting import plot_decision_regions


value = 1.80
width = 0.90

clf1 = LogisticRegression(random_state=42)
clf2 = DecisionTreeClassifier(random_state=42) 
clf3 = MLPClassifier(random_state=42, verbose = 0)
clf4 = RandomForestClassifier(random_state=42)
eclf = EnsembleVoteClassifier(clfs=[clf4,], weights=[1, 1, 1, 1], voting='soft')

X_list = MiceImputed[["Sunshine", "Humidity9am", "Cloud3pm"]] #took only really important features
X = np.asarray(X_list, dtype=np.float32)
y_list = MiceImputed["RainTomorrow"]
y = np.asarray(y_list, dtype=np.int32)

# Plotting Decision Regions
gs = gridspec.GridSpec(3,3)
fig = plt.figure(figsize=(18, 14))

labels = ['Logistic Regression',
          'Decision Tree',
          'Neural Network',
          'Random Forest']

for clf, lab, grd in zip([clf1, clf2, clf3, clf4, eclf],
                         labels,
                         itertools.product([0, 1, 2],
                         repeat=2)):
    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf, 
                                filler_feature_values={2: value}, 
                                filler_feature_ranges={2: width}, 
                                legend=2)
    plt.title(lab)

plt.show()

```

```python
# In summary, the code generates a combined visualization where the time taken for execution is represented as bars on one 
# y-axis, and the accuracy scores are represented as lines on another y-axis. This allows you to compare the models' 
# accuracy and execution time in a single graph. The two y-axes share the same x-axis, which represents different machine
# learning models.

accuracy_scores = [accuracy_lr, accuracy_dt, accuracy_nn, accuracy_rf ]
roc_auc_scores = [roc_auc_lr, roc_auc_dt, roc_auc_nn, roc_auc_rf ]
coh_kap_scores = [coh_kap_lr, coh_kap_dt, coh_kap_nn, coh_kap_rf ]
tt = [tt_lr, tt_dt, tt_nn, tt_rf, ]

model_data = {'Model': ['Logistic Regression','Decision Tree','Neural Network','Random Forest',],
              'Accuracy': accuracy_scores,
              'ROC_AUC': roc_auc_scores,
              'Cohen_Kappa': coh_kap_scores,
              'Time taken': tt}
data = pd.DataFrame(model_data)

fig, ax1 = plt.subplots(figsize=(12,10))
ax1.set_title('Model Comparison: Accuracy and Time taken for execution', fontsize=13)
color = 'tab:green'
ax1.grid()
ax1.set_xlabel('Model', fontsize=13)
ax1.set_ylabel('Time taken', fontsize=13, color=color)
ax2 = sns.barplot(x='Model', y='Time taken', data = data, palette='summer')
ax1.tick_params(axis='y')
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Accuracy', fontsize=13, color=color)
ax2 = sns.lineplot(x='Model', y='Accuracy', data = data, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
```

```python
#  create a comparative visualization of different models based on two evaluation metrics: Area Under the Receiver Operating 
# Characteristic Curve (ROC_AUC) and Cohen's Kappa. Let's break down what each part of the code does:

# combines a bar plot (using blue bars) for ROC_AUC values and a line plot (using red lines) for Cohen's
# Kappa values to compare different models. It utilizes two y-axes to represent the two different evaluation metrics on the 
# same plot, allowing you to visually compare the performance of each model with respect to both metrics simultaneously.

fig, ax3 = plt.subplots(figsize=(12,10))
ax3.set_title('Model Comparison: Area under ROC and Cohens Kappa', fontsize=13)
color = 'tab:blue'
ax3.grid()
ax3.set_xlabel('Model', fontsize=13)
ax3.set_ylabel('ROC_AUC', fontsize=13, color=color)
ax4 = sns.barplot(x='Model', y='ROC_AUC', data = data, palette='winter')
ax3.tick_params(axis='y')
ax4 = ax3.twinx()
color = 'tab:red'
ax4.set_ylabel('Cohen_Kappa', fontsize=13, color=color)
ax4 = sns.lineplot(x='Model', y='Cohen_Kappa', data = data, sort=False, color=color)
ax4.tick_params(axis='y', color=color)
plt.show()
```
